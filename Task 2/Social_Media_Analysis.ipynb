```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')

df['Timestamp'] = pd.to_datetime(df['Timestamp'])

def preprocess_text(text):
    text = text.lower()
    text = ''.join([char for char in text if char.isalnum() or char.isspace()])
    return text

df['processed_text'] = df['Text'].apply(preprocess_text)

analyzer = SentimentIntensityAnalyzer()
def get_sentiment_score(text):
    scores = analyzer.polarity_scores(text)
    return scores['compound']

df['sentiment_score'] = df['processed_text'].apply(get_sentiment_score)

sentiment_trends = df.groupby(pd.Grouper(key='Timestamp', freq='ME'))['sentiment_score'].mean()

plt.figure(figsize=(10, 6))
plt.plot(sentiment_trends.index, sentiment_trends.values)
plt.xlabel('Time')
plt.ylabel('Mean Sentiment Score')
plt.title('Sentiment Trend Over Time')
plt.grid(True)
plt.show()
```

```python
hashtags_data = {
    'Platform': ['Twitter', 'Twitter', 'Instagram', 'Instagram', 'Facebook', 'Facebook'],
    'Hashtag': ['#DataScience', '#MachineLearning', '#DataScience', '#AI', '#DeepLearning', '#Python'],
    'Count': [100, 150, 80, 120, 90, 70]
}
hashtag_df = pd.DataFrame(hashtags_data)

hashtag_trends = hashtag_df.groupby(['Platform', 'Hashtag'])['Count'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Hashtag', y='Count', hue='Platform', data=hashtag_trends)
plt.xlabel('Hashtag')
plt.ylabel('Count')
plt.title('Hashtag Trends on Each Platform')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
platform_data = {
    'Platform': ['Twitter', 'Facebook', 'Instagram', 'YouTube', 'TikTok'],
    'Reach': [1000, 800, 700, 900, 1200],  
    'Engagement': [0.15, 0.10, 0.12, 0.08, 0.20]  
}
platform_df = pd.DataFrame(platform_data)

platform_df['Impact_Score'] = platform_df['Reach'] * platform_df['Engagement']
plt.figure(figsize=(10, 6))
sns.barplot(x='Platform', y='Impact_Score',color='r' , data=platform_df)
plt.xlabel('Platform')
plt.ylabel('Impact Score (Reach * Engagement)')
plt.title('Platform Impact on Public Opinion')
plt.show()
most_significant_platform = platform_df.loc[platform_df['Impact_Score'].idxmax()]
print(f"\nThe platform with the most significant impact is: {most_significant_platform['Platform']} (Impact Score: {most_significant_platform['Impact_Score']})")
```

```python
platform_engagement = {
    'Platform': ['Twitter', 'Facebook', 'Instagram', 'YouTube', 'TikTok'],
    'Likes': [1500, 2000, 1800, 2500, 3000],
    'Retweets/Shares': [800, 1200, 900, 1500, 2000]
}

engagement_df = pd.DataFrame(platform_engagement)

plt.figure(figsize=(10, 6))

plt.bar(engagement_df['Platform'], engagement_df['Likes'], label='Likes', width=0.4, color='skyblue')

plt.bar(engagement_df['Platform'], engagement_df['Retweets/Shares'], label='Retweets/Shares', width=0.4, color='lightcoral', bottom=engagement_df['Likes'])

plt.xlabel('Platform')
plt.ylabel('Number of Interactions')
plt.title('Comparison of Likes and Retweets/Shares Across Platforms')
plt.legend()
plt.tight_layout()
plt.show()

engagement_df['Total_Interaction'] = engagement_df['Likes'] + engagement_df['Retweets/Shares']
most_popular_platform = engagement_df.loc[engagement_df['Total_Interaction'].idxmax()]
print(f"\nThe platform with the highest total interactions is: {most_popular_platform['Platform']} (Total Interactions: {most_popular_platform['Total_Interaction']})")
```
